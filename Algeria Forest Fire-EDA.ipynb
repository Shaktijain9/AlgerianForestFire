### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

### Loading the dataset

```python
df = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv", )
```

#### Checking first 5 rows of the dataset

```python
df.head()
```

<b>OBSERVATION : </b>   Since the dataset contains Bejaia Region data & Sidi-Bel Abbes Region Data we're going to read it separately.

#### Reading the Bejaia Region data from CSV file.

```python
bejaia_df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', skiprows=1, nrows=122)
bejaia_df.head()
```

### Addition of Region Column

```python
bejaia_df["Region"] = "Bejaia"
bejaia_df
```

```python
bejaia_df
```

#### Reading the Sidi-Bel Abbes Region data from CSV file.

```python
abbes_df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', skiprows=126)
abbes_df.head()
```

### Addition of Region Column

```python
abbes_df['Region'] = "Sidi-Bel Abbes"
abbes_df
```

### Merging these two dataframes into one

```python
df = abbes_df.append(bejaia_df).reset_index()
df
```

### Shape of the dataframe

```python
df.shape
```

### Information of the dataframe

```python
df.info()
```

<b>OBSERVATION : </b> Features DC & FWI are string objects hence we need to convert them into float values.

### Feature datatypes

```python
df.dtypes
```

### Dropping irrelavant column

```python
df.drop(columns=['index'], inplace=True)
```

```python
df.head()
```

### Columns in dataframe

```python
df.columns
```

### Checking the null values

```python
df.isna().sum()
```

<b>OBSERVATION : </b> There is one null value present in the Classes feature.

#### Dropping the row which has missing value

```python
df.dropna(inplace=True)
df.isna().sum()
```

### Checking the shape of the df after dropping the row

```python
df.shape
```

#### Changing the name of  features 

```python
df.rename(columns={'Classes  ': "Classes", 'Rain ': "Rain", " Ws": "Wind-Speed", " RH":"Relative Humidity"}, inplace=True)
```

### Creation of date column from [day, month, year] features

```python
df['date']= df.loc[:, 'day':'year'].apply(lambda x:"/".join(x.astype('str')), axis=1)
```

```python
df
```

### Dropping columns [day, month, year] from dataset

```python
df.drop(columns=['day', 'month', 'year'], inplace=True)
```

#### Converting date string to datetime object

```python
df['date'] = pd.to_datetime(df['date'])
```

#### Converting the type for Feature DC

```python
df["DC"] = df['DC'].map(lambda x: str(x).replace(" ", "")).astype('float')
```

#### Converting the type for Feature FWI

```python
df['FWI'] = df['FWI'].astype('float')
```

#### Checking Values for Classes Feature

```python
df['Classes'] = df['Classes'].str.strip()
```

```python
df["Classes"].value_counts()
```

```python
df
```

### Statistical Summary of the data

```python
df.describe()
```

#### Making Date as index for the dataframe

```python
df.set_index('date', inplace=True)
```

### Univariate Analysis

#### Distribution plot for all the numerical Values

```python
feature_filter = df.columns[(df.dtypes=="int64") | (df.dtypes=='float64')]
```

```python
pos=1
fig = plt.figure(figsize=(13,25))

for i in feature_filter:
    ax = fig.add_subplot(8,3,pos)
    pos = pos + 1
    sns.distplot(df[i], ax=ax, color="#e39d07")
```

#### Box plot distribution for all the numeric features

```python
sns.set_style("whitegrid")
plt.figure(figsize=(16, 8))
sns.boxplot(data=df)
```

#### Count Matrix of Region wrt Classes

```python
pd.crosstab(df['Classes'], df['Region'], normalize=True)*100
```

<b>OBSERVATION : </b> There are more fire incidents happened in Sidi-Bel Abbes region as compared to Bejaia region.

```python

sns.countplot(data=df, x="Classes")
```

<b>OBSERVATION : </b> High rainfall months were June & September.

#### Wind-Speed wrt monthly data

```python
temp = df['Wind-Speed'].resample('M')
temp.plot(ylabel = "Wind-Speed (km/h)", xlabel="Date")
```

<b>OBSERVATION : </b> Windspeed was high during the month of October.

#### Relative Humidity wrt monthly data

```python
temp = df['Relative Humidity'].resample('M')
temp.plot(ylabel = "Relative Humidity (%)", xlabel="Date")
```

<b>OBSERVATION : </b> Humidity seems high in the months of August, September & October
